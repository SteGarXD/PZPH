```python
import requests
import pandas as pd
import psycopg2
```

```python
books_url = "https://www.anapioficeandfire.com/api/books"
response_books = requests.get(books_url)
books_data = response_books.json()

df_books = pd.DataFrame(books_data)
df_books.head()
```

```python
houses = []
page = 1

while True:
    url = f"https://www.anapioficeandfire.com/api/houses?page={page}&pageSize=50"
    response = requests.get(url)
    data = response.json()
    if not data:
        break
    houses.extend(data)
    page += 1

df_houses = pd.DataFrame(houses)
df_houses.head()
```

```python
houses_with_titles_and_words = []
page = 1

while True:
    url = f"https://www.anapioficeandfire.com/api/houses?page={page}&pageSize=50&hasTitles=true"
    response = requests.get(url)
    data = response.json()

    if not data:
        break

    houses_with_titles_and_words.extend(data)
    page += 1

houses_with_titles_and_words = [
    house for house in houses_with_titles_and_words
    if house.get("words") and house["words"].strip() != ""
]

houses_df = pd.DataFrame(houses_with_titles_and_words)
houses_df.head()
```
