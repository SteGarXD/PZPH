```python
class Chocolate:
    def __init__(self, price: float, add_on: str = None):
        if price <= 0:
            raise ValueError("Цена батончика должна быть больше нуля.")
        self.price = price
        self.add_on = add_on

    def show_my_chocolate(self):
        if self.add_on:
            print(f"Шоколадный батончик, добавка: {self.add_on}")
        else:
            print("Обычный шоколадный батончик")

    def __str__(self):
        return f"Шоколадный батончик, {self.price} рублей"

#Примеры объектов с аргументами:
try:
    choco1 = Chocolate(34.0)
    choco2 = Chocolate(56.0, "Нуга и арахис")
    choco3 = Chocolate(115.0, "Изюм")
#1
    choco1.show_my_chocolate()
    choco2.show_my_chocolate()
    choco3.show_my_chocolate()
#2
    print(choco1)
    print(choco2)
    print(choco3)
#3
    choco4 = Chocolate(-20.0)
except ValueError as e:
    print(e)
```
